**Importing the neccsary Libaries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading the dataset and exploring**

```python
data = pd.read_csv('Train_data.csv')
data.head(10)
```

```python
data.info()
```

**Dropping some Useless Columns**

```python
data.drop('Customer Id', axis=1, inplace=True)
```

**Encoding the categorical values so the model can Understand the data**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data['Building_Painted']=le.fit_transform(data['Building_Painted'])
data['Building_Fenced']=le.fit_transform(data['Building_Fenced'])
data['Garden']=le.fit_transform(data['Garden'])
data['Settlement']=le.fit_transform(data['Settlement'])
```

**Filling the missing data with Mean, Median, or Mode**

```python
median_date= data['Date_of_Occupancy'].median()
data['Date_of_Occupancy'].fillna(median_date, inplace=True) 
```

```python
mode_geo= data['Geo_Code'].mode()[0]
data['Geo_Code'].fillna(mode_geo, inplace=True)
```

```python
median_dimension= data['Building Dimension'].median()
data['Building Dimension'].fillna(median_dimension, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
data.info()
```

**Converting the data types of the columns to numeric**

```python
data['NumberOfWindows']= pd.to_numeric(data['NumberOfWindows'], errors='coerce')
data['NumberOfWindows'].fillna(data['NumberOfWindows'].median(), inplace=True)
data['Geo_Code']= pd.to_numeric(data['Geo_Code'], errors='coerce')
data['Geo_Code'].fillna(data['Geo_Code'].mode()[0], inplace=True)
```

```python
data.info()
```

**Splitting the data into Samples and Features and Training Them**

```python
x=data.drop('Claim', axis=1)
y=data['Claim']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size=0.3, random_state=42)
```

**Scaling the Features Using Standard Scaler**

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

**Importing My Model**

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(x_train, y_train)
```

**Predicting the test data**

```python
y_pred= model.predict(x_test)
print(f"Actual Predction: {y_test}, Model Predictions: {y_pred}")
```

**Checking the model perfomance using Metrics**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy= accuracy_score(y_test, y_pred)
confusion= confusion_matrix(y_test, y_pred)
report= classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{confusion}")
print(f"Classification Report:\n{report}")
```

The **Accuracy** is 78% but the **Recall** is 14% which shows that my model performed poorly in terms of indicating Claim cases
